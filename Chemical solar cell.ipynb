# Chemical solar cell example from Chapter 6

Ansatz.  
Starts with cell at equilibrium.  The species are $\ce{H2O(l)}$, $\ce{H2(g)}$, and $\ce{O2(g)}$ linked by the reversible reaction:

$$\ce{H2O(l) <=>[G][R] H2(g) + 1/2 O2(g) }$$

At equilibrium, the forward ($G$, generation) and reverse ($R$, recombination) rates of the above reaction are equal. 

Note: due to the finite the vapor pressure of $\ce{H2O}$, $\ce{H2O(g)}$ would be present, as well, but this is not a part of the analysis.  

To make a connection to solar energy conversion, the following assumptions are made. 

* It is assumed that photons in can split water into hydrogen and oxygen.  In this case, there is extra generation, $\Delta G$, beyond the equilbrium rate, and thus the $\ce{H2}$ and $\ce{O2}$ partial pressures will rise above their equilibrium values. 
* The cell can have product-selective membranes which can remove $\ce{H2(g)}$ from the cell in one place and $\ce{O2(g)}$ in another. The flux across these membranes is controlled by the pressure of the product outside of the cell. That is, gas will be transmitted through the membrane if its pressure is higher inside the cell than outside.  In this way, one can imagine converting the absorbed photons which drove the water splitting reaction into chemical energy, the separated $\ce{H2}$ and $\ce{O2}$ gas products. 

The following two limiting cases are discussed.
* If $\ce{H2}$ and $\ce{O2}$ are not removed, the pressure will increase until reverse reaction, making water, is equal to the generation rate. When this steady state pressure is reached, $\Delta R = \Delta G$. There is an analogy here to the open circuit voltage of a solar cell.   
* $\ce{H2}$ and $\ce{O2}$ are removed from the cell by the membranes, so that their recombination rate is same as equilibrium. This is analogous to the short circuit current of a solar cell. No chemical energy is transferred (system cannot perform work) because very low pressures of  $\ce{H2}$ and $\ce{O2}$ on the other sides of the membranes would need to be maintained.

We are interested in what happens between these two cases: $\Delta R = \Delta G$ (open circuit) and $\Delta R = 0$ (short circuit).  
We can suppose that the steady state partial pressures are inversely proportional to the recombination rate in the gas phase. For example, if we were to add a catalyst such as Pt to the gas phase side, the recombination rate would be increase and the pressure would not go up as much for a given $\Delta G$.  

In the rest of the example, the rates will be scaled to $R_{\ce{H2}}^0$, the recombination rate of hydrogen in equilibrium. In a solar cell, we actually know what the minimum value of the recombination is, namely the radiative rate need to maintain thermal generation. But here it is used a normalization constant.   

The free energy balance ($F$ is free energy, not $G$ to avoid confusion with generation) is shown below.  
Free energy: adding/subtracting molecules (also electrons and holes) at their chemical potential.

$$dF = \mu_{\ce{H2}} dN_{\ce{H2}} + \mu_{\ce{O2}} dN_{\ce{O2}} + \mu_{\ce{H2O}} dN_{\ce{H2O}}  $$

We also know the reaction stoichiometry

$$dN_{\ce{H2}} = 2dN_{\ce{O2}} = -dN_{\ce{H2O}}$$

At equilibrium (free energy of products = free energy of reactants) 

$$ \mu_{\ce{H2}} + 1/2 \mu_{\ce{O2}} = \mu_{\ce{H2O}} $$

Clearly, at one bar of the gases and liquid water at standard conditions (298 K and 1 bar pressure for the gases) the system is not at equilibrium as the chemical potentials of the gases are both zero and the chemical potential of liquid water is -2.46 eV. 

To solve for the *equilbrium* gas partial pressures we use the dependence of the chemical potential on pressure (note that here the superscript zero means standard conditions).  

$$\mu_{\ce{H2}} = \mu_{\ce{H2}}^0 + k_B T \log(p/p_0)$$

$$\mu_{\ce{O2}} = \mu_{\ce{O2}}^0 + k_B T \log(p/p_0)$$

This allows a calculation of the $\ce{H2}$ pressure at equilibrium for 0.2 bar of $\ce{O2}$ (atmospheric concentration). 

$$ k_B T \log(p_{\ce{H2}}/p_0) + 1/2 \, k_B T \log(0.2) = \mu_{\ce{H2O}}^0 $$

$$ \log(p_{\ce{H2}}/p_0) = \frac{\mu_{\ce{H2O}}^0}{k_B T} - 1/2\log(0.2)$$  

The equilibrium $\ce{H2}$ pressure is tiny but not zero (~1e-41 bar).

```python
# Zero here means standard conditions
Î¼_H2_gas_zero = 0*eV
Î¼_O2_gas_zero = 0*eV
Î¼_H2O_liquid_zero = -2.46*eV   # 2x the water splitting E0

# Chemical potential depends on pressure for a gas.  
def Î¼_H2(p, T=300*kelvin, p_0=1.):
    return Î¼_H2_gas_zero + kB*T*np.log(p/p_0)

def Î¼_O2(p, T=300*kelvin, p_0=1.):
    return Î¼_O2_gas_zero + kB*T*np.log(p/p_0)

# Equilibrium calculations
p_O2_equil = 0.2
p_H2_equil = np.exp(Î¼_H2O_liquid_zero/(kB*300*kelvin) - np.log(p_O2_equil)/2)
print('Equilibrium H2 pressure is {:.3e} bar for an O2 pressure of 0.2 bar'.format(p_H2_equil))

Î¼_H2_equil = Î¼_H2(p_H2_equil)
print('Equilibrium H2 chemical potential is {:.3f} eV'.format(Î¼_H2_equil/eV))
```

This situation with $p_{\ce{O2}}$ >> $p_{\ce{H2}}$ is analogous to a p-type semiconductor at equilibrium with a large excess of holes.   

Next, we consider deviations from equilibrium and the chemical driving forces which attempt to restore it.  
The following equations will be parametric in the equilibrium recombination rate.  
The equilibrium partial pressure of $\ce{H2}$ is very small.  But if there is even less $\ce{H2}$ for some reason, then the recombination reaction would need to slow down with respect to the generation rate to restore equilbrium. Thinking about the reaction orders, recombination rate for hydrogen obviously needs $\ce{H2}$ but  the generation rate does not, as it uses water which is in infinite supply in this example.  

The recombination reaction is first-order in $p_{\ce{H2}}$ and half-order in $p_{\ce{O2}}$.

$$R_{\ce{H2}} \sim p_{\ce{H2}} p_{\ce{O2}}^{1/2}$$  

The recombination rate at arbitrary pressures for $\ce{H2}$ and $\ce{O2}$ can be related to the rate at equilbrium 

$$R_{\ce{H2}} = R_{\ce{H2}}^0 
\frac{ p_{\ce{H2}}} { p^0_{\ce{H2}}}
\sqrt{\frac{ p_{\ce{O2}}} { p^0_{\ce{O2}}} } $$  

where the zero superscripts mean equilbrium. One can see that increasing $p_{\ce{H2}}$ above its equilbrium value will increase the recombination rate. This is also true for $p_{\ce{O2}}$ but the increase will be less due to the square root dependence due to the reaction order. 

Now, we consider the effect of light to create extra generation $\Delta G_{\ce{H2}}$. We will focus on the effects of light generation on the pressure and chemical potential of $\ce{H2}$ but note that a similar analysis could be done for $\ce{O2}$. By mass balance, in steady state, the $\ce{H2}$ current out of the cell is $R-G$ (current defined to be negative here for a reduction product).  

$$j_{\ce{H2}} = 2 j_{\ce{O2}} = R_{\ce{H2}} - G_{\ce{H2}} $$ 

We can express $ R_{\ce{H2}}$ in terms of the gas partial pressures referenced to their equilibrium values and expand the generation term into the sum of the equilibrium and light-driven values:

$$j_{\ce{H2}} = 2 j_{\ce{O2}} =  R_{\ce{H2}}^0 \frac{ p_{\ce{H2}} \sqrt{p_{\ce{O2}}}}{p_{\ce{H2}}^0 \sqrt{{p^0_{\ce{O2}}}}} -G^0_{\ce{H2}} - \Delta G_{\ce{H2}}  $$

Finally, noting that $R_{\ce{H2}}^0  = G_{\ce{H2}}^0 $ at equilibrium, we get the $\ce{H2}$ current as a function of the gas pressures and light generation rate only. 

$$j_{\ce{H2}} = 2 j_{\ce{O2}} =  R_{\ce{H2}}^0 \left(\frac{ p_{\ce{H2}} \sqrt{p_{\ce{O2}}}}{p_{\ce{H2}}^0 \sqrt{{p^0_{\ce{O2}}}}} - 1\right) - \Delta G_{\ce{H2}}  $$

Key step is knowing that the recombination rates (now kinetics enters) are proportional to pressure.  And pressure is related to $\mu$ in equilibrium.   Recombination rate is linear in the H2 pressure and expontential in the H2 chemical potential.  

In the following, he assumes values for $R_{\ce{H2}}^0$ and scales generation to it. 

```python
def j_H2_pressure(p_H2, RzeroH2=1, ÎGH2=0, p_O2=p_O2_equil):
    term1 = (p_H2/p_H2_equil)
    term2 = (p_O2/p_O2_equil)**0.5
    return RzeroH2*(term1*term2 - 1) - ÎGH2
    
p_H2_norm = np.linspace(0.01, 3, 50)

fig = plt.figure(figsize=(4.5,4), dpi=100)
plt.title('H$_2$ current vs. pressure\n $p_{O_2}$ = $p^0_{O_2}$', y = 1.02)
plt.plot(p_H2_norm, j_H2_pressure(p_H2_norm*p_H2_equil), 'r', label='dark')
plt.plot(p_H2_norm, j_H2_pressure(p_H2_norm*p_H2_equil, ÎGH2=1), 'g', label='light, $\Delta G = 1$')
plt.xlim(0,3)
plt.xlabel('$p_{H_2}}$/$p^0_{H_2}}$')
plt.ylabel('$j_{H_2}$/$R^0_{H_2}$')

plt.legend()
plt.grid()
plt.axvline(x=1, linestyle='--', color='lightslategrey')
plt.axhline(y=0, linestyle='--', color='lightslategrey')

plt.gca().set_facecolor('aliceblue')

plt.show()
```

Red curve shows that changing $\ce{H2}$ pressure will lead to current in the dark.  
Green curve shows that a scaled light generation rate of 1 will lead to a current of 1 at short circuit or a doubling of $p_{\ce{H2}}$ at open circuit.   

Now use the relation between gas pressure and chemical potential to express the hydrogen current as a function of $\mu_{\ce{H2}}$  

$$j_{\ce{H2}} = R_{\ce{H2}}^0 \left[ \exp\left( \frac{\mu_{\ce{H2}}+ 1/2 \, \mu_{\ce{O2}}+ \mu_{\ce{H2O}}}{k_B T} \right) - 1\right]  - \Delta G_{\ce{H2}} $$

The chemical potential of water (-2.46 eV) will not change as we assume it is in infinite supply.  
In principle, the chemical potention of $\ce{O2}$ will change, but only by tiny amounts because it is in large excess compared to $\ce{H2}$  
So this can be reexpressed in terms of the change in the chemical potential of $\ce{H2}$ in the gas phase:  

$$j_{\ce{H2}} = R_{\ce{H2}}^0 \left[ \exp\left( \frac{\Delta\mu_{\ce{H2}}}{k_B T} \right) - 1\right]  - \Delta G_{\ce{H2}} $$

This looks very much like the diode equation, which was the point of the example. 
> $\Delta G_{\ce{H2}}$ is short circuit current.   
> Value of $\Delta\mu_{\ce{H2}}$ which leads to zero current is open circuit voltage

Overview of this thought experimnent.   
* There are 2 mobile species in equilibrium with an infinite reservoir.  $R_0$ and $G_0$ are equal and maintain the two species in equilibrium.  
* Add additional generation $\Delta G$. There has to be additional recombination $\Delta R$ to restore equilibrium if you turn off the light.  
* The minority species can flow out of the system, if the $\mu$ on the other side is low enough.  The $\mu$ that stops the current depends on the ratio of the pump generation to the equilibrium recombination rate

$$\Delta\mu^{OC}_{\ce{H2}} = k_B T \log \left(\frac{\Delta G_{\ce{H2}}}{R_{\ce{H2}}^0} + 1 \right)$$

This is analogous to open circuit voltage of a solar cell:
